```python
import numpy as np
import matplotlib.pyplot as plt
# Where is the dataset saved
root = "C:/Users/nightfury/Desktop/PYTHON BASICS/"

root_Parking = f"{root}CBuilding/csv/c_building_pedestrian_clear_anon/"
dir_velodyne = f"{root_Parking}/velodyne/"
velodyne_data_list = []

# Load data into velodyne_data_list
for i in range(499):
    file_path = f"{dir_velodyne}/{i:06d}.csv"
    data = np.loadtxt(file_path, usecols=(0, 1, 2))
    velodyne_data_list.append(data)

# Process data and plot histogram
all_data = np.concatenate(velodyne_data_list)  # Concatenate all data arrays into one
GG1 = np.square(all_data[:, 0])
HH1 = np.square(all_data[:, 1])
II1 = np.square(all_data[:, 2])

FF1 = (GG1 + HH1 + II1)
KK1 = np.sqrt(FF1)
print(all_data)
bins = np.linspace(0,50,50)
plt.hist(KK1, histtype='bar', bins=bins)
plt.ylabel('Cloudpoints')
plt.xlabel('Distance in meters')
plt.title('Velodyne Histogram (fog)')
plt.legend()
plt.yscale('linear')
plt.show()
```
